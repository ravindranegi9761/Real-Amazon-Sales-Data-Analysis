# PRELIMINARY ANALYSIS

```python
t_r=amazon.groupby('Month')["Order_amount(₹)"].sum()
print(t_r)
print('\n')
a_j=t_r['April']-t_r['June']
t_r_p=(a_j/t_r['April'])*100
print(f' decreased revenue from april to june :{t_r_p:.2f}')
print('\n')
d_r_a_m=t_r['April']-t_r['May']
d_r_a_m_p=(d_r_a_m/t_r['April'])*100
print(f'decreased revenue from april to may :{d_r_a_m_p:.2f}')
print('\n')
d_r_m_j=t_r['May']-t_r['June']
d_r_m_j_p=(d_r_m_j/t_r['May'])*100
print(f' decreased revenue from may to june : {d_r_m_j_p:.2f}')
```

```python
r_s=amazon.groupby('Category')['Order_amount(₹)'].sum()
a=r_s.sum()
s_r_a=r_s['Set']/a*100
print(f'total set revenue :{s_r_a:.2f}')
k_r_a=r_s['kurta']/a*100
print('\n')
print(f'total kurta revenue :{k_r_a:.2f}')
print('\n')
wd_r_a=r_s['Western Dress']/a*100
print(f'total western dress revenue :{wd_r_a:.2f}')
```

```python
t_r=amazon.groupby(['Category'],as_index=False)['order_amount_($)'].mean().sort_values(by='order_amount_($)',ascending=False)
t_v = t_r.sort_values(by='order_amount_($)', ascending=False).head()
t_v
```

```python
a=amazon.groupby('Customer_type')['order_amount_($)'].mean()
b=a['Business']
print("average order amount by customer_type")
print(f'business: ${b:.2f}')
c=a['Customer']
print(f'customer: ${c:.2f}')
```

```python
cancelled_orders_count = ((amazon['Status'] == 'Cancelled')|(amazon['Status'] == 'Shipped - Lost in Transit')).sum()
cancelled_orders_count
```

```python
return_order_count=((amazon['Status']=='Shipped - Returned to Seller') | 
                    (amazon['Status']=='Shipped - Returning to Seller')|
                    (amazon['Status']=='Shipped - Rejected by Buyer')).sum()
return_order_count
```

```python
total_orders = amazon.shape[0]
```

```python
print("Total Cancelled Order = ",cancelled_orders_count)
print("Total order= ",total_orders)
p_o_c=cancelled_orders_count*100/total_orders
print("percent of cancelled order= ",round(p_o_c,2))
r_o=return_order_count*100/total_orders
print("percent of returned order= ",round(r_o,2))
c_a_r=cancelled_orders_count+return_order_count
print("total cancelled and returned order =",c_a_r)
percent_of_c_r_o=round(c_a_r*100/amazon['Qty'].sum(),2)
print("percent of cancelled and returned order= ",percent_of_c_r_o)
```
